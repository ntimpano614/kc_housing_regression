```python
from bs4 import BeautifulSoup 
import pandas as pd
import requests
```

```python
url = 'https://www.ciclt.net/sn/clt/capitolimpact/gw_ziplist.aspx?FIPS=53033'
r = requests.get(url)
c=r.content
soup = BeautifulSoup(c,'html.parser')
```

```python
first = soup.find_all('tr')
```

```python
len(first)
```

```python
for x in first:
    if x.find('tr'):
        table_row = x.find('tr')
        try:
            holder = table_row.find('td').text
        except:
            print('not found')
#         print(table_row.find('td').text)


zip_lst = holder.split('\n')

for i in zip_lst:
    if i == 'King County':
        zip_lst.remove(i)
```

```python
zip_lst
```

```python
import re 
regexp = re.compile(r'\D+')

for index, i in enumerate(zip_lst):
    if '9' in i:
        if regexp.search(i):
            numeric_filter = filter(str.isdigit,i)
            numeric_string = ''.join(numeric_filter)
            zip_lst[index] = numeric_string
            print(numeric_string)
            print(i)
```

```python
zip_lst.pop()
zip_lst
```

```python
# grab all the evens and put into zips
# grab all odds and put into towns
# add to dataframe and export 
zip_dict = {}
for index, element in enumerate(zip_lst):
    if index % 2 == 0:
        # zips
        if element in zip_dict.keys():
            zip_dict[element].append(zip_lst[index+1])
        else:
            zip_dict[element] = [zip_lst[index+1]] 
```

```python
zips_df = pd.DataFrame(list(zip_dict.items()),columns=['zip','towns'])
zips_df
```

```python
zips_df.to_csv('zips_kc.csv', index=False)
```
